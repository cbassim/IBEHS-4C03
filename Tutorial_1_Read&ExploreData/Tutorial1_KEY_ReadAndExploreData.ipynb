## Tutorial Step 7: Filter the dataframe (all columns) that have less than 1650 mils for position 1:   pos1 < 1650


```python
# Get all columns (the whole dataframe) with rows which have pos1 < 1650 mils
# Method 1 Square brackets column indexing 
less_1650_1 = boards_no_time[boards_no_time["Pos1"] < 1650]
print(less_1650_1)
```

```python
# Get all columns (the whole dataframe) with rows which have pos1 < 1650 mils
# Method 2 is using the location labeling of the dataframe. 
# Using the .loc can enable you to select and slice more effeciently  
less_1650_2 = boards_no_time.loc[boards_no_time['Pos1'] < 1650]
print(less_1650_2)
```

```python
# Get all columns (the whole dataframe) with rows which have pos1 < 1650 mils
# Method 3 Use of the pandas .query function
less_1650_3=boards_no_time.query('Pos1 < 1650')
print(less_1650_3)
```

## Tutorial Step 8: Create a Visualization for board thickness by position

Think of the data type for board thickness and for what you want to communicate with your plot:  create a bar graph and a boxplot. What can you say about the relationship between board thickness and board position?

We will review some Principles of the Graphical Excellence, but number 1 is to always make sure that you are labeling your plots.  

import and use the plotting libraries, either matplotlib or seaborn.  

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mean_mils=boards_no_time.mean()
print(mean_mils)
```

```python
plt.bar(x=["Pos1","Pos2","Pos3","Pos4","Pos5","Pos6"],height=mean_mils,color='black')
plt.xlabel("Position", size=14)
plt.ylabel("Thickness (mils)", size=14)
```

```python
# Seaborn!.  Note the standard deviation added.
sns.set(rc = {'figure.figsize':(10,6)})
sns.set_style("white")
barplotC = sns.barplot(data=boards_no_time, color="gray", errorbar='sd')
barplotC.set_xticklabels(barplotC.get_xticklabels(), rotation=45)
barplotC.set_xlabel("Position", size=14)
barplotC.set_ylabel("Thickness (mils)", size=14)
barplotC.set_title("Boards Thickness by Position", size=14)
```
